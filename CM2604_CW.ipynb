```python
import pandas as pd
```

```python
train_path = r'C:\Users\nisal\Machine Learning\adult\adult.test'
test_path = r'C:\Users\nisal\Machine Learning\adult\adult.data'
```

```python
# view test data
df1 = pd.read_csv(test_path)
df1.head()
```

```python
# view train data
df2 = pd.read_csv(train_path)
df2.head()
```

```python
columns = [
    "age", "workclass", "final-weight", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", 
    "capital-loss", "hours-per-week", "native-country", "income"]
```

```python
# assigning columns
# remove white space after every ','
# show '?' values as 'NaN'
train_df = pd.read_csv(test_path, names=columns, sep=',\s', na_values=["?"],engine='python')
train_df.head()
```

```python
# assigning columns
# remove white space after every ','
# show '?' values as 'NaN'
# remove first row 
test_df = pd.read_csv(train_path, names=columns, sep=',\s', na_values=["?"], skiprows=1,engine='python')
test_df.head()
```

```python
# check data types before merge
print("TRAIN DATA TYPES")
print(train_df.dtypes)
print("------------------------")
print("TEST DATA TYPES")
print(test_df.dtypes)
```

```python
# train data types and test data types equal
# merge two datasets
merged_df = pd.concat([train_df, test_df])
merged_df.head()
```

```python
merged_df.info()
```

```python
from sklearn.utils import shuffle
```

```python
# shuffle merged data for prevents any bias and ensure data is randomized
shuffled_df = shuffle(merged_df)
shuffled_df.head()
```

```python
# save shuffled data as new csv
dir = r'C:\Users\nisal\Machine Learning\census_income_datset.csv'
# without index as a separate column
shuffled_df.to_csv(dir, index=False)
print('Saved at',dir)
```

```python
census_df = pd.read_csv(dir)
census_df.head()
```

```python
census_df.info()
```

```python
census_df.nunique()
```

```python
# check unique values in each columns
for column in census_df:
    print(column)
    print(census_df[column].unique())
    print('\n')
```

```python
# replace values in 'income' column
census_df['income'] = census_df['income'].replace({'<=50K.': '<=50K', '>50K.': '>50K'})
```

```python
# check updated unique values in each columns
for column in census_df:
    print(column)
    print(census_df[column].unique())
    print('\n')
```

```python
# check null(nun) values
nun_values = census_df.isnull().sum()
total_nun_values = nun_values.sum()
print(nun_values)
print("total null(nun) values :",total_nun_values)
```

```python
# nan values replace from forward filling
census_df.fillna(method = 'ffill', inplace = True)
```

```python
# check null(nun) values again
nun_values = census_df.isnull().sum()
total_nun_values = nun_values.sum()
print(nun_values)
print("total null(nun) values :",total_nun_values)
```

```python
# check how many duplicate rows
duplicate_rows = census_df.duplicated().sum()
print("Number of duplicate rows :", duplicate_rows)
```

```python
# keep means drop duplicates without first occurrence.
census_df.drop_duplicates(keep='first', inplace=True)
```

```python
# check how many duplicate rows again
duplicate_rows = census_df.duplicated().sum()
print("Number of duplicate rows :", duplicate_rows)
```
